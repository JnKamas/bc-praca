```python
names = ["100k-2023.csv", "250k-2023.csv", "500k-2023.csv", "1m-2023.csv", "2m-2023.csv", "4_5m-2023.csv"]
sizes = [100000, 250000, 500000, 1000000, 2000000, 4500000]
```

```python
plt.figure(figsize=(10, 6))

x = np.arange(0, 4)
y = 3 * x

for name, size in zip(names, sizes):

    df = pd.read_csv("./vis_data/unweighted-vis-"+name)

    x_values = (df['samples'] / size) * 100
    y_values = df['diff']


    plt.plot(x_values, y_values, '-', markersize=1, label=f'{size:,} voters')


plt.plot(x, y, label='Linear trend')
plt.xlabel('Size of coordinated group (% from all voters)')
plt.ylabel('Change of mandates')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 4)
y = 3 * x

plt.figure(figsize=(10, 6))

for name, size in zip(names, sizes):

    df = pd.read_csv("./vis_data/unweighted-vis-"+name)

    x_values = (df['samples'] / size) * 100
    y_values = df['diff']


    plt.plot(x_values, y_values - (3 * x_values), '-', markersize=1, label=f'{size:,} voters')

plt.plot(x, y * 0, label='Linear trend')
plt.xlabel('Size of coordinated group (% from all voters)')
plt.ylabel('Change of mandates')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 4)
y = 3 * x

plt.figure(figsize=(10, 6))

size = 1_000_000
file_name = names[sizes.index(size)] 

df_unweighted = pd.read_csv("./vis_data/unweighted-vis-" + file_name)
x_values_unweighted = (df_unweighted['samples'] / size) * 100
y_values_unweighted = df_unweighted['diff']

plt.plot(x_values_unweighted, y_values_unweighted - (3 * x_values_unweighted), '-', markersize=1, label=f'Unweighted {size:,} voters')

df_weighted = pd.read_csv("./vis_data/weighted-vis-" + file_name)
x_values_weighted = (df_weighted['samples'] / size) * 100
y_values_weighted = df_weighted['diff']

plt.plot(x_values_weighted, y_values_weighted - (3 * x_values_weighted), '-', markersize=1, label=f'Weighted {size:,} voters')

# Linear trend
plt.plot(x, y * 0, label='Linear trend')

plt.xlabel('Size of coordinated group (% from all voters)')
plt.ylabel('Change of mandates')

plt.legend()
plt.show()
```
