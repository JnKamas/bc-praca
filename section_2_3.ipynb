```python
plt.figure(figsize=(10, 6))

x = np.arange(0, 4)
y = 3 * x

for name, size in zip(names, sizes):

    df = pd.read_csv("./vis_data/unweighted-vis-"+name)

    x_values = (df['samples'] / size) * 100
    y_values = df['diff']


    plt.plot(x_values, y_values, 'o', markersize=1, label=f'{size} voters')


plt.plot(x, y, label='Linear trend')
plt.xlabel('Size of coordinated group (% from all voters)')
plt.ylabel('Change of mandates')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 4)
y = 3 * x

plt.figure(figsize=(10, 6))

for name, size in zip(names, sizes):

    df = pd.read_csv("./vis_data/unweighted-vis-"+name)

    x_values = (df['samples'] / size) * 100
    y_values = df['diff']


    plt.plot(x_values, y_values - (3 * x_values), 'o', markersize=1, label=f'{size} voters')

plt.plot(x, y * 0, label='Linear trend')
plt.xlabel('Size of coordinated group (% from all voters)')
plt.ylabel('Change of mandates')
# plt.title('Comparison of different sizes')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

x = np.arange(0, 4)
y = 3 * x

size = 1000000

files = ["unweighted-vis-1m-2023","weighted-vis-1m-2023", "electable-unweighted-vis-1m-2023", "electable-weighted-vis-1m-2023"]
names = ["Unweighted","Weighted by number of votes", "Only electable parties, unweighted", "Electable parties, weighted"]

for file, name in zip(files, names):

    df = pd.read_csv(f"./vis_data/{file}.csv")

    x_values = (df['samples'] / size) * 100
    y_values = df['diff']

    plt.plot(x_values, y_values - (3 * x_values), 'o', markersize=1, label=name)


plt.plot(x, y * 0, label='Linear trend')

plt.xlabel('size of coordinated group (% from all voters)')
plt.ylabel('change of seats')
plt.title('Comparison of different averages, v = 1,000,000')

plt.legend()
plt.show()
```
